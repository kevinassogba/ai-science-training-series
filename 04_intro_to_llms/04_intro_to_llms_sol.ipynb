## Homework Solutions

### 1- Playing with generative parameters

```python
selected_model = "meta-llama/Llama-3.2-1B"
input_prps = ["My dog wanted to eat ice cream because",
              "You are helpful to humanity when",
              "Charity begins at",
              "Early to bed, early to"]
generator = pipeline("text-generation", model=selected_model, device_map="auto")
```

#### Configuration 1

```python
generator(input_prps, max_length=20, num_return_sequences=5)
```

#### Configuration 2

```python
generator(input_prps, max_length=15, num_return_sequences=5)
```

#### Configuration 3

```python
generator(input_prps, max_length=15, num_return_sequences=5, temperature=1.3)
```

#### Configuration 4

```python
generator(input_prps, max_length=15, num_return_sequences=5, temperature=0.1)
```

#### Configuration 5

```python
new_generator = pipeline("text-generation", model="gpt2", device_map="auto")
new_generator(input_prps, max_length=15, num_return_sequences=5, temperature=0.1)
```

#### Observations:

Varying the max_length, temperature, and model leads to the following observations:

* The max_length value defines how long the model output is. If the max_length is too high, the model may continue padding more unnecessary and irrelevant text after the correct output.
* The temperature contributes to the stability of model output across the number of return sequences. By default, the temperature is 1, and a smaller value (e.g., 0.1) leads to the same text for all return sequences. Playing with a higher temperature (e.g., 1.3) yielded more variations across outputs.
* The model itself is critical to the quality of the generated text. Comparing GPT-2  with "meta-llama/Llama-3.2-1B" showed that Llama-3.2-1B generated more accurate/expected text than GPT-2. This performance benefit can for example be attributed to the number of parameters (117 million for GPT-2 vs 1 billion for Llama-3.2-1B).

### 2- BertViz analysis

```python
from transformers import AutoTokenizer, AutoModel, utils, AutoModelForCausalLM

from bertviz import model_view
utils.logging.set_verbosity_error()  # Suppress standard warnings

def visualize(model_name, input_txt):
    model = AutoModelForCausalLM.from_pretrained(model_name, output_attentions=True)
    tokenizer = AutoTokenizer.from_pretrained(model_name)
    inputs = tokenizer.encode(input_txt, return_tensors='pt')
    outputs = model(inputs)
    attention = outputs[-1]
    tokens = tokenizer.convert_ids_to_tokens(inputs[0])
    model_view(attention, tokens)
```

#### Model = openai-community/gpt2

```python
input_text = "Early to bed, early to"
model_name = 'openai-community/gpt2'
visualize(model_name, input_text)
```

#### Model = meta-llama/Llama-3.2-1B

```python
model_name = 'meta-llama/Llama-3.2-1B'
visualize(model_name, input_text)
```

#### Observations:

Comparing openai-community/gpt2 and meta-llama/Llama-3.2-1B shows that the attention mechanism is more diversified with more layers, e.g., Layer 9 - Head 12 or Layer 15 for Llama-3.2-1B. In contrast, almost every layer in GPT-2 focuses on the first token "Early". Nevertheless, the overall trend is similar for both models in that they pay more attention to the beginning of the text. A key difference between the two models is that Llama-3.2-1B prepends a token <|begin_ot_text}>.
